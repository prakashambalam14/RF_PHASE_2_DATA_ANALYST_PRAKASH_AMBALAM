```python
import pandas as pd
```

```python
df=pd.read_csv("video_game_sales.csv")
df.head()
```

```python
df.shape
```

```python
df = df.drop_duplicates()
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['last_update'] = pd.to_datetime(df['last_update'], errors='coerce')
```

```python
df["developer"].isnull().sum()
```

```python
df['developer'] = df['developer'].fillna('Unknown')
```

```python
df["developer"].isnull().sum()
```

```python
df["critic_score"].isnull().sum()
```

```python
df['critic_score'].fillna(df['critic_score'].median(), inplace=True)
```

```python
df["critic_score"].isnull().sum()
```

```python
sales_cols = ['total_sales','na_sales','jp_sales','pal_sales','other_sales']
df[sales_cols] = df[sales_cols].fillna(0)
```

```python
#Which games generated the highest total sales?

top_games = df.sort_values('total_sales', ascending=False)[
    ['title','console','genre','publisher','total_sales']
].head(10)

top_games
```

```python
#Which genres generate the most revenue?
genre_sales = df.groupby('genre')['total_sales'] \
                .sum() \
                .sort_values(ascending=False)

genre_sales.head(10)
```

```python
#Which consoles produced the highest total sales?
console_sales = df.groupby('console')['total_sales'] \
                  .sum() \
                  .sort_values(ascending=False)

console_sales.head(10)
```

```python
#Which publishers dominate total game sales?

publisher_sales = df.groupby('publisher')['total_sales'] \
                    .sum() \
                    .sort_values(ascending=False)

publisher_sales.head(10)
```

```python
#Which region contributes the most to global sales?

region_totals = df[['na_sales','jp_sales','pal_sales','other_sales']].sum()
region_totals
```

```python
#Which genre performs best in each region?

region_genre = df.groupby('genre')[['na_sales','jp_sales','pal_sales','other_sales']] \
                 .sum() \
                 .sort_values('na_sales', ascending=False)

region_genre.head(10)
```
